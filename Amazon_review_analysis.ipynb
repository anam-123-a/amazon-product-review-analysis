Analyze Amazon product review data to find:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("amazon.csv")
print(df)
```

```python
print("Total Products:", len(df))
```

```python
average_rating = df["rating"].mean()

print("Average Rating:", round(average_rating, 2))
```

```python
print("Average Rating:", round(df["rating"].mean(),2))
```

```python
rating_distribution = df["rating"].value_counts().sort_index()

rating_distribution
```

```python
plt.figure(figsize=(8,5))

plt.bar(rating_distribution.index, rating_distribution.values)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Products")

plt.show()
```

```python
top_rated = df.sort_values(by="rating", ascending=False)

top_rated[["product_name","rating"]].head(10)
```

### Top Rated Products

- Several products received a perfect **5.0-star rating**.
- Products such as **LED Bulb**, **Coffee Mug**, **Wireless Mouse**, and **Bluetooth Speaker** are among the highest-rated products.
- These products indicate excellent customer satisfaction.

```python
most_reviewed = df.sort_values(by="rating_count", ascending=False)

most_reviewed[["product_name","rating_count"]].head(10)
```

### Most Reviewed Products

- Bluetooth Speaker 29 has the highest number of reviews (4942).
- Notebook and Backpack products also received a large number of customer reviews.
- Products with more reviews are generally more popular among customers.

```python
rating_distribution = df["rating"].value_counts().sort_index()

rating_distribution
```

```python
rating_distribution = df["rating"].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.bar(
    rating_distribution.index.astype(str),
    rating_distribution.values,
    width=0.6
)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Products")

plt.xticks(rotation=45)

plt.show()
```

### Insight

- The highest number of products have a rating of **4.2** (8 products).
- **7 products** have a perfect rating of **5.0**.
- Very few products have ratings below **3.0**.
- Most products are rated between **3.5 and 5.0**, indicating generally positive customer satisfaction.

```python
plt.figure(figsize=(10,5))

plt.bar(
    most_reviewed["product_name"].head(10),
    most_reviewed["rating_count"].head(10)
)

plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Product Name")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    top_rated["product_name"].head(10),
    top_rated["rating"].head(10)
)

plt.title("Top Rated Products")
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.xticks(rotation=45)

plt.show()
```
